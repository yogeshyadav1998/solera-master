```python
import pandas as pd
import re
```

```python
df=pd.read_csv('processed_merged.csv')
```

```python
df['search_salts']=df['salt'].str.lower()
```

```python
df['salt'][0]
```

```python
df['search_salts']=df['search_salts'].str.replace('(','')
df['search_salts']=df['search_salts'].str.replace(')','')
df['search_salts']=df['search_salts'].str.replace('.','')
```

```python
df['search_salts']=df['search_salts'].str.replace('[0-9]+iu|[0-9]+ml|[0-9]+gm|[0-9]+%|[0-9]+mg|[0-9]+mcg|[0-9]+ppm','')
df['search_salts']=df['search_salts'].str.replace('[0-9]+\siu|[0-9]+\sml|[0-9]+\sgm|[0-9]+\s%|[0-9]+\smg|[0-9]+\smcg|[0-9]+\sppm|\s/','')
```

```python
df['search_salts']=df['search_salts'].str.replace('w/w|w/v|%|ppm|[0-9]+','')
```

```python
df['search_salts']=df['search_salts'].str.replace('+',',')
```

```python
df['search_salts']=df['search_salts'].str.replace(' ','')
```

```python
df.to_csv('processed_merged.csv',index=False)
```
